```python
import cubewalkers as cw
from cana.datasets.bio import load_all_cell_collective_models
```

```python
def cell_collective_models():
    return [cw.Model(cw.conversions.network_rules_from_cana(BN)) 
            for BN in load_all_cell_collective_models()]
```

```python
W = 1000
T = 2000
```

```python
loaded_models = cell_collective_models()
```

```python
total_models = len(loaded_models)
for model_idx,model in enumerate(loaded_models):
    model.n_walkers = W
    model.n_time_steps = T
    N = model.n_variables
    
    print(f"Simulating Model {model_idx} ({W=},{T=},{N=}). . .")
    model.simulate_ensemble(averages_only=True,threads_per_block=(16,16))
    print(f"Progress: {(model_idx+1)/total_models}")
```
